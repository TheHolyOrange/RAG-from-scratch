# Importing Packages

```python
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np
```

# Sample Documents

```python
documents = [
    "Students must maintain at least 75% attendance to be eligible for exams.",
    "Hostel gates close at 9 PM on weekdays and 10 PM on weekends.",
    "The refund deadline for semester fees is 15 days after the start of classes."
]
```

# Embeddings

```python
model = SentenceTransformer("all-MiniLM-L6-v2")

doc_embeddings = model.encode(documents)

print(doc_embeddings.shape)
```

```python
print(doc_embeddings)
```

# FAISS

```python
dimension = doc_embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
```

```python
index.add(np.array(doc_embeddings, dtype="float32"))
```

# Ask a Question

```python
query = "What attendance is required for exams?"
```

```python
query_embedding = model.encode([query])
```

# Search top matching documents

```python
k = 2
distances, indices = index.search(np.array(query_embedding, dtype="float32"), k)
```

```python
print(indices)
print(distances)
```

```python
print("Query:", query)
print("\nTop retrieved documents:\n")

for rank, i in enumerate(indices[0], start=1):
    print(f"Rank {rank}: {documents[i]}")
    print(f"Distance: {distances[0][rank-1]:.4f}\n")
```
